```python
# The Central Limit Theorem
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
population = np.random.exponential(scale=2, size=100000)
```

```python
sample_means = [np.mean(np.random.choice(population, 50)) for _ in range(1000)]
```

```python
plt.hist(sample_means, bins=50)
plt.title("CLT in action — means of exponential samples")
plt.show()
```

```python
print(f"Mean of sample means: {np.mean(sample_means):.3f}")
```

```python
print(f"Std of sample means: {np.std(sample_means):.3f}")
```

```python
# Hypothesis Testing
```

```python
from scipy import stats
import numpy as np
```

```python
control   = np.random.binomial(1, 0.10, 1000)
treatment = np.random.binomial(1, 0.13, 1000)
```

```python
t_stat, p_value = stats.ttest_ind(control, treatment)
```

```python
print(f"Control mean:   {control.mean():.3f}")
print(f"Treatment mean: {treatment.mean():.3f}")
print(f"T-statistic:    {t_stat:.3f}")
print(f"P-value:        {p_value:.4f}")
```

```python
if p_value < 0.05:
    print("Reject H0 — statistically significant difference")
else:
    print("Fail to reject H0 — insufficient evidence")
```
