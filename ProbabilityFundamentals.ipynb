```python
#Conditional Probability
```

```python
# Given information
P_disease = 0.01           # 1% of people have disease
P_healthy = 0.99           # 99% are healthy
```

```python
P_positive_if_disease = 0.99    # Test catches 99% of sick people
P_positive_if_healthy = 0.01    # Test falsely flags 1% of healthy people
```

```python
# Let's test 10,000 people to visualize:
total_people = 10000
```

```python
# How many actually have disease?
people_with_disease = total_people * P_disease  # 100 people
```

```python
# How many are healthy?
people_healthy = total_people * P_healthy  # 9,900 people
```

```python
# Of sick people, how many test positive?
sick_test_positive = people_with_disease * P_positive_if_disease  # 99
```

```python
# Of healthy people, how many test positive (false alarms)?
healthy_test_positive = people_healthy * P_positive_if_healthy  # 99
```

```python
# Total positive tests
total_positive = sick_test_positive + healthy_test_positive  # 198
```

```python
# Of all positive tests, how many are actually sick?
P_disease_given_positive = sick_test_positive / total_positive
print(f"P(Disease|Positive) = {P_disease_given_positive:.1%}")  # 50%!
```

```python
#Normal Distribution (Gaussian)
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate normal distribution
mean = 100      # Average
std_dev = 15    # Spread (standard deviation)
```

```python
data = np.random.normal(mean, std_dev, 10000)
```

```python
# Plot
plt.hist(data, bins=50, density=True, alpha=0.7)
plt.axvline(mean, color='red', linestyle='--', label=f'Mean = {mean}')
plt.axvline(mean + std_dev, color='orange', linestyle='--', label=f'±1 SD')
plt.axvline(mean - std_dev, color='orange', linestyle='--')
plt.legend()
plt.title('Normal Distribution (IQ Scores)')
plt.xlabel('IQ Score')
plt.ylabel('Frequency')
plt.show()
```

```python
# What percentage of people have IQ > 130?
from scipy.stats import norm
```

```python
# IQ 130 is 2 standard deviations above mean
# From 68-95-99.7 rule: 95% are within ±2 SD
# So 5% are outside, and half of those (2.5%) are above

p_above_130 = 1 - norm.cdf(130, loc=100, scale=15)
print(f"P(IQ > 130) = {p_above_130:.2%}")  # ≈ 2.3%
```

```python
#Binomial Distribution
```

```python
from scipy.stats import binom
```

```python
# Flip coin 10 times, what's probability of exactly 7 heads?
n = 10      # trials
p = 0.5     # fair coin
k = 7       # successes we want
```

```python
prob = binom.pmf(k, n, p)
print(f"P(exactly 7 heads in 10 flips) = {prob:.2%}")  # ≈ 11.7%
```

```python
# What about 7 or more heads?
prob_7_or_more = 1 - binom.cdf(6, n, p)  # cdf gives P(X ≤ 6)
print(f"P(7+ heads) = {prob_7_or_more:.2%}")  # ≈ 17.2%
```

```python
#Poisson Distribution
```

```python
from scipy.stats import poisson
```

```python
# Average 5 customers per hour
lambda_ = 5
```

```python
# What's probability of exactly 3 customers in next hour?
prob_3 = poisson.pmf(3, lambda_)
print(f"P(exactly 3 customers) = {prob_3:.2%}")  # ≈ 14%
```

```python
# What's probability of 8 or more customers?
prob_8_or_more = 1 - poisson.cdf(7, lambda_)
print(f"P(8+ customers) = {prob_8_or_more:.2%}")  # ≈ 13%
```
