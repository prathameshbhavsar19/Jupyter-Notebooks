Dot Product

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create two vectors
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])
```

```python
# Three ways for Dot Product

r1 = np.dot(a, b)
r2 = a @ b
r3 = np.sum(a * b)
```

```python
print(r1, r2, r3)
```

Matrix Multiplication

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])
```

```python
print(A.shape)
print(A.T.shape)
print(A.T)
```

```python
A_inv = np.linalg.inv(A)
```

```python
result = A @ A_inv
print(np.round(result, 10))
```

```python
AB = A @ B
BA = B @ A
```

```python
print("A @ B =", AB)
print("B @ A =", BA)
print("Are they equal?", np.array_equal(AB, BA))
```

```python
points = np.array([[0, 0],
                   [1, 0],
                   [1, 1],
                   [0, 1]])
```

```python
print(points.shape)
```

```python
angle_deg = 45
angle_rad = np.radians(angle_deg)
```

```python
R = np.array([[np.cos(angle_rad), -np.sin(angle_rad)],
               [np.sin(angle_rad), np.cos(angle_rad)]])
```

```python
R_points = points @ R.T
```

```python
print(R_points)
```

Plot these points

```python
plt.scatter(points[:, 0],   points[:, 1],   label='Points',         color='blue')
plt.scatter(R_points[:, 0], R_points[:, 1], label='Rotated Points', color='red')
plt.axis('equal')
plt.legend()
plt.show()
```

```python
plt.fill(points[:, 0],   points[:, 1],   alpha=0.3, color='blue', label='Original')
plt.fill(R_points[:, 0], R_points[:, 1], alpha=0.3, color='red',  label='Rotated')
plt.axis('equal')
plt.legend()
plt.grid(True)
plt.show()
```
