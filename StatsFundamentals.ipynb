```python
# Hypothesis Testing
```

```python
import numpy as np
from scipy import stats
```

```python
# Simulated test scores
caffeine_scores = [78, 85, 82, 90, 88, 76, 92, 84, 87, 89,
                   81, 86, 83, 91, 79, 88, 85, 87, 82, 90,
                   84, 88, 86, 89, 91, 83, 87, 85, 88, 90]
```

```python
placebo_scores = [75, 78, 72, 80, 76, 74, 82, 77, 79, 81,
                  73, 79, 75, 83, 71, 80, 76, 78, 74, 82,
                  77, 79, 78, 81, 83, 75, 80, 76, 79, 82]
```

```python
print(f"Caffeine group mean: {np.mean(caffeine_scores):.2f}")
print(f"Placebo group mean: {np.mean(placebo_scores):.2f}")
```

```python
#State Hypotheses
#H₀: Caffeine has NO effect on test scores (μ_caffeine = μ_placebo)
#H₁: Caffeine DOES affect test scores (μ_caffeine ≠ μ_placebo)
```

```python
t_statistic, p_value = stats.ttest_ind(caffeine_scores, placebo_scores)
```

```python
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4f}")
```

```python
# Data
n = 100           # flips
observed = 60     # heads we got
p_null = 0.5      # probability under H₀
```

```python
# Expected under null
expected = n * p_null  # 50
```

```python
# Standard error
se = np.sqrt(n * p_null * (1 - p_null))
print(f"Standard Error: {se}")  # 5.0
```

```python
# Z-score (how many standard errors away from expected?)
z = (observed - expected) / se
print(f"Z-score: {z}")  # 2.0
```

```python
# P-value: probability of getting 60+ heads if coin is fair
p_value = 1 - stats.norm.cdf(z)  # Area to the right of z=2.0
print(f"P-value: {p_value:.4f}")  # 0.0228 (2.28%)
```

```python
alpha = 0.05
```

```python
if p_value < alpha:
    print("Reject H₀: Coin is likely BIASED")
else:
    print("Fail to reject H₀: Not enough evidence of bias")

# Result: 0.0228 < 0.05, so we REJECT H₀
# Conclusion: Evidence suggests coin is biased toward heads
```

```python
#One-Sample T-Test
```

```python
# Sample data: heights of 30 people
np.random.seed(42)
heights = np.random.normal(175, 10, 30)  # mean=175, sd=10, n=30
```

```python
# Hypotheses:
# H₀: μ = 170 (true mean is 170)
# H₁: μ ≠ 170 (true mean is different)
```

```python
# Calculate t-test
t_statistic, p_value = stats.ttest_1samp(heights, 170)
```

```python
print(f"Sample mean: {heights.mean():.2f} cm")
print(f"T-statistic: {t_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
```

```python
if p_value < 0.05:
    print("\n✓ REJECT H₀: Mean height is significantly different from 170cm")
else:
    print("\n✗ FAIL TO REJECT H₀: Not enough evidence of difference")
```

```python
# Two groups
male_heights = np.random.normal(175, 10, 30)
female_heights = np.random.normal(165, 10, 30)
```

```python
# Hypotheses:
# H₀: μ_male = μ_female
# H₁: μ_male ≠ μ_female
```

```python
# Two-sample t-test
t_stat, p_value = stats.ttest_ind(male_heights, female_heights)
```

```python
print(f"Male mean: {male_heights.mean():.2f} cm")
print(f"Female mean: {female_heights.mean():.2f} cm")
print(f"Difference: {male_heights.mean() - female_heights.mean():.2f} cm")
print(f"P-value: {p_value:.4f}")
```

```python
if p_value < 0.05:
    print("\n✓ Significant difference between groups")
else:
    print("\n✗ No significant difference")
```
